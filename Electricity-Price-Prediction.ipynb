```python
#Veri tiplerini dntren fonksiyon tanmland
def convert_data_types(df, convert_from, convert_to):
    cols = df.select_dtypes(include=[convert_from]).columns
    for col in cols:
        df[col] = df[col].values.astype(convert_to)
    return df
```

```python
def algo_test(x, y):
    
    L = LinearRegression()
    E = ElasticNet()
    R = Ridge()
    Lass = Lasso()
    ETR = ExtraTreeRegressor()
    GBR = GradientBoostingRegressor()
    XGBC = XGBRegressor()
    RFR =RandomForestRegressor()
    
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.15, random_state=42)
    
    algos = [L, E, R, Lass, ETR, GBR, XGBC,RFR]
    algo_names= ['Linear', 'ElasticNet', 
                 'Ridge', 'Lasso', 
                 'Extra Tree', 
                 'Gradient Boosting', 
                 'XGBRegressor','Random Forest']
    r_squared = []
    rmse = []
    mae = []
    
    result = pd.DataFrame(columns = ['R_Squared', 'RMSE', 'MAE'],
                           index = algo_names)
    
    for algo in algos:
        algo.fit(x_train, y_train)
        algo.predict(x_test)
        
        r_squared.append(r2_score(y_test, algo.predict(x_test)))
        rmse.append(mean_squared_error(y_test, algo.predict(x_test))**.5)
        mae.append(mean_absolute_error(y_test, algo.predict(x_test)))
        
    result.R_Squared = r_squared
    result.RMSE = rmse
    result.MAE = mae
    
    return result.sort_values('R_Squared', ascending=False)
```

```python
algo_test(x, y)
```
